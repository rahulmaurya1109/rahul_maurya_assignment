```python
#ANSWER 9. The method as as below

#right-justifying  ==> rjust()
#left-justifying   ==> ljust()
#centering         ==> center()

strr = " my name is rahul "
print(strr.rjust(50,'#'))  #rjust() ex.
print(strr.ljust(50,'#'))  #ljust() ex.
print(strr.center(50,'#')) #cente() ex.
```

```python
#ANSWER 10.
# there is three function which is use for the removing the whitespace from start, end or from the both.
#(1) lstrip()
#(2) rstrip()
#(3) strip()

d ="      my name is rahul     "
print(d.lstrip())  # use to remove whitespace from left
print(d.rstrip())  # use to remove whitespace from right
print(d.strip())   # use to remove whitespace from left and right
```
