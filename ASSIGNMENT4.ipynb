```python
#ANSWER 11. there is del , remove(), pop(), clear() are the type which can use to remove the items from the list.

val = [1,2,3,4,5]
del val[0]
print(val)
#del will use to remove item by the index value.

val.remove(2)
print(val)
#remove() is use to remove the value the name.

val.pop()
print(val)
# pop() is use to remove last item of the list.

val.clear()
print(val)
# clear() is use to remove all items from the list 
```

```python
# ANSWER 17. in cop.copy() the old and new list object is sharing the same memory location so if we change into the any of list the changes will
#appers in both list. but in the copy.deep.copy() the object is not have the same memory location any changes in any list will not appers in other list.

# copy.copy() is do shallow copy of the list.
# in normal list shallow copy
import copy
list1 = [1,2,3,4,5]
list2 = copy.copy(list1)
print(list2)
list2[2] = 100
print(list2)
print(list1)

# in nested list shallow copy
list3 = [[1,2,3],[4,5,6],[7,8,9]]
list4 = copy.copy(list3)
print(list3)
print(list4)
list4[1][0] = 100
print(list4)
print(list3)


# deepcopy() in normal list 
import copy
list1 = [1,2,3,4,5]
list2 = copy.deepcopy(list1)
print(list2)
list2[2] = 100
print(list2)
print(list1)

#deepcopy() in the nested list
list3 = [[1,2,3],[4,5,6],[7,8,9]]
list4 = copy.deepcopy(list3)
print(list3)
print(list4)
list4[1][0] = 100
print(list4)
print(list3)
```
