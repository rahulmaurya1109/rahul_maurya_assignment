```python
#ANSWER9.

x = 10 #global variable 
def func():
    x = 5 #local variable 
    print("the local variable is ", x)
func()

k = 10
def glb():
    global k
    k = 15
    print("the local variable is", k)
glb()
print("the global variable is now changed ",k)


#AS PER THE GIVEN EXAMPLE ABOVE WE CAN REFER THE FUCNTION VARIABLE TO THE GLOBAL VARIBALE 
#BY USING THE GLOBAL KEYWORD TO DECLARE IT TO THE GLOBAL VARIABLE 
```
